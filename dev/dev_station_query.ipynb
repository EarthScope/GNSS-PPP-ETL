```python
import datetime
import logging

from gnss_product_management import GNSSClient

logging.basicConfig(
    level=logging.INFO,
    format="%(levelname)-8s %(name)s  %(message)s",
)
```

```python
date = datetime.datetime(2025, 1, 1, tzinfo=datetime.timezone.utc)
client = GNSSClient.from_defaults()

lat = 57.22003
lon = -153.27918
radius_m = 100000
# ── Build the query ───────────────────────────────────────────────────────
sq = (
    client.station_query()
    .centers("ERT")
    .within(lat, lon, radius_m / 1000)
    .rinex_version("3")
    .on(date)
)
```

```python
# ── 2. Remote search ─────────────────────────────────────────────────────
print("\n── Remote RINEX search ──────────────────────────────────")
results = sq.search()
for r in results:
    print(f"  {r.parameters.get('SSSS', '?'):>4s}  v{r.parameters.get('V', '?')}  {r.uri}")
if not results:
    print("  (no remote files found)")
```
